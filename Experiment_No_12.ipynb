```python
import pandas as pd
data={
    "Roll No":[1,2,3,4,5],
    "Name":["A","B","C","D","E"],
    "Age":[20,"-",22,23,"24"],
    "Marks":[90,91,92,93,94],
    "Department":["CSE","IT","CSE","CSE","ETC"],
    "Admission_Date":["01-06-2023","03-06-2023","05-06-2023","07-06-2023","16-06-2023"]
}
df=pd.DataFrame(data)
df.to_csv("data.csv",index=False)
df=pd.read_csv("data.csv")
print(df)
```

```python
df.replace("-",np.nan,inplace=True)
#convert colums to numeric where needed
df["Age"]=pd.to_numeric(df["Age"],errors="coerce")
df["Marks"]=pd.to_numeric(df["Marks"],errors="coerce")
df
```

```python
#step 3 CHeck missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
#fill missing range  with mean
df["Age"].fillna(df["Age"].mean(),inplace=True)
df
```

```python
#step 5: Fix inconsistent text formatting
df["Department"]=df["Department"].str.upper()
df
```

```python
#step 6: Convert Admission_Date to standard date time
df["Admission_Date"]=pd.to_datetime(df["Admission_Date"],errors="coerce")
df
```

```python
# step 7: cleanded dataset
df.to_csv("cleaned_data.csv",index=False)
df
```

```python
# step 8: save cleaned dataset
df.to_csv("cleaned_student_data.csv",index=False)
print("\nCleaned CSV file saved as cleaned_student_data.csv ")
```
